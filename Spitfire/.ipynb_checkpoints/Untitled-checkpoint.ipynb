```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
path = os.getcwd() + '/data_input/2021-3-25/gold/0_gold_test_ppp_1s_bin8/run1'
```

```python
files = sorted([i for i in os.listdir(path) if '.txt' in i and 'ps' not in i and 'powers' not in i])

data = [i for i in files if '_bg' not in i]
data_indices = [str(i) for i in np.arange(len(data))]

bkg = [i for i in files if '_bg' in i]
bkg_indices = [str(i) for i in np.arange(len(bkg))]

col_names = ['wl','1','2','cts']

ef=pd.DataFrame()
for i in bkg:
    i_name = i.rstrip('.txt')
    df = pd.read_csv(path+'/'+i, sep='\t', names=col_names)
    df.drop(columns=['1','2'], inplace=True)
    df['wn'] = 10000000/df['wl'] - 12579
    df.sort_values(by='wn',inplace=True)
    df.columns = pd.MultiIndex.from_product([[i_name], df.columns])
    df = df.reset_index(drop=True)
    ef = pd.concat([ef,df],axis=1)

ef.loc[:,ef.columns.get_level_values(1)=='wn'].head(50)
```

```python
def read_files(path):
    files = sorted([i for i in os.listdir(path) if '.txt' in i and 'ps' not in i and 'powers' not in i])
    
    data = [i for i in files if '_bg' not in i]
    data_indices = [str(i) for i in np.arange(len(data))]
    
    bkg = [i for i in files if '_bg' in i]
    bkg_indices = [str(i) for i in np.arange(len(bkg))]
    
    col_names = ['wl','1','2','cts']
    
    df_data = pd.DataFrame()
    for i in data:
        df = pd.read_csv(path+'/'+i, sep='\t', names=[i.rstrip('.txt')+'wl','1','2',i.rstrip('.txt')+'cts'])
        df.drop(columns=['1','2'])
        df[i.rstrip('.txt')+'wn'] = 10000000/df[i.rstrip('.txt')+'wl'] - 12579
        df.sort_values(by=i.rstrip('.txt')+'wn',inplace=True)
        pd.merge(df_data,df,how='right')
        
    return df_data
read_files(path)
```
